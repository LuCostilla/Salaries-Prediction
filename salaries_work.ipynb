```python
lst_con= []


for pais in t_data.company_location:
    if pais in AMN_ME:
        lst_con.append('AMN_ME')
    if pais in EUR:
        lst_con.append('EUR')
    if pais in ME_2:
        lst_con.append('ME_2')    
    if pais in AFR:
        lst_con.append('AFR')
    if pais in AM_S:
        lst_con.append('AM_S')    
    if pais in APAC:
        lst_con.append('APAC')   
    if pais in OCE:
        lst_con.append('OCE')
        
```
